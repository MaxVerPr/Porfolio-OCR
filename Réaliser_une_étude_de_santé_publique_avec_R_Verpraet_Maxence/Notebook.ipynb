```python
#Afficher les dimensions du dataset
nutrition.shape
```

```python
#Consulter le nombre de colonnes
print("Le tableau nutrition comporte {} colonne(s)".format(nutrition.shape[1]))
```

```python
#Afficher les 5 premières lignes de la table
nutrition.head()
```

```python
#Conversion de la colonne sous nutrition en numérique
nutrition['Valeur'] = pd.to_numeric(nutrition['Valeur'], errors='coerce')
```

```python
#Conversion de la colonne (avec l'argument errors=coerce qui permet de convertir automatiquement les lignes qui ne sont pas des nombres en NaN)
#Puis remplacement des NaN en 0
nutrition['Valeur'] = pd.to_numeric(nutrition['Valeur'], errors='coerce')
nutrition = nutrition.replace(np.nan, 0)
```

```python
#changement du nom de la colonne Valeur par sous_nutrition
nutrition.rename(columns={'Valeur':'Sous_nutrition'}, inplace=True)
```

```python
#Multiplication de la colonne sous_nutrition par 1000000
nutrition['Sous_nutrition'] = nutrition['Sous_nutrition']*1000000
```

```python
#Afficher les 5 premières lignes de la table
nutrition.head()
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.1 - Proportion de personnes en sous nutrition</h3>
</div>

```python
# Il faut tout d'abord faire une jointure entre la table population et la table sous nutrition, en ciblant l'année 2017
pers_s_nut = pd.merge(population,nutrition, on='Zone', how='inner')
pers_s_nut = pers_s_nut[(pers_s_nut['Année_x'] == 2017) & (pers_s_nut['Année_y'] == '2016-2018')]
```

```python
#Affichage du dataset
pers_s_nut.head()
```

```python
#Calcul et affichage du nombre de personnes en état de sous nutrition
nombre_sous_nutrition = pers_s_nut['Sous_nutrition'].sum()
population_total = pers_s_nut['Population'].sum()
prop_sous_nutrition = nombre_sous_nutrition / population_total * 100
print('Le % de personne en état de sous nutrition est de :',prop_sous_nutrition)
print('Le nombre de personne en sous nutrition est de',nombre_sous_nutrition)
print('Le nombre de personne dans le monde est de',population_total)
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Nombre théorique de personne qui pourrait être nourries</h3>
</div>

```python
#Combien mange en moyenne un être humain ? Source => 
#2500 kcal, source entre 2250 et 2750kcal
#https://fr.wikipedia.org/wiki/Ration_alimentaire
```

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
df = pd.merge(population, dispo, on='Zone', how='inner')
df = df[(df['Année'] == 2017)]
```

```python
#Affichage du nouveau dataframe
df.head()
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
df['dispo_kcal'] = df.iloc[:,2]*df.iloc[:,7]
df
```

```python
#Calcul du nombre d'humains pouvant être nourris
calcul_humains = (df['dispo_kcal'].sum()/2500)
print("Nombre d'humains pouvant être nourris :",calcul_humains)
```
